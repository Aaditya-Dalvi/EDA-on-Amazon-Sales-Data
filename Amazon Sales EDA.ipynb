```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('amazon.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

# Data Cleaning

```python
df.isnull().sum()
```

Droping some columns

```python
df.drop(columns=['product_id','user_id','user_name','review_id','img_link','product_link'],inplace=True)
df.head()
```

```python
df = df.dropna(axis=0)
```

Cleaning the features like **discounted_price, actual_price, discount_percentage, rating_count** and **rating** and converting them to correct types for further analysis.

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','')
df['discounted_price'] = df['discounted_price'].astype(float)

df['actual_price'] = df['actual_price'].str.replace('₹','').str.replace(',','')
df['actual_price'] = df['actual_price'].astype(float)

df['discount_percentage'] = df['discount_percentage'].str.replace('%','')
df['discount_percentage'] = df['discount_percentage'].astype(int)

df['rating_count'] = df['rating_count'].str.replace(',','')
df['rating_count'] = df['rating_count'].astype(int)

df['rating'] = df['rating'].str.replace('|', '')
df['rating'] = df['rating'].replace('', np.nan)
df['rating'] = df['rating'].astype(float)

df.head()
```

```python
df.info()
```

# Creating Some New Features

```python
df_cat = df['category'].str.split('|', expand=True).rename(columns={0:'category_1', 1:'category_2', 2:'category_3', 3:'category_4',4:'category_5'})
 
df['category_1'] = df_cat['category_1']
df['category_2'] = df_cat['category_2']
df['category_3'] = df_cat['category_3']
df['category_4'] = df_cat['category_4']

df.drop('category', axis=1, inplace=True)
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['Price'] = df['discounted_price']*df['rating_count']
df['Price']
```

# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Count of each Category

```python
# Get the count of each category and sort in descending order
category_counts = df['category_2'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.countplot(x='category_2', data=df, order=category_counts.index, palette='viridis')
plt.title('Distribution of Category 2 ')
plt.xlabel('Category')
plt.ylabel('Count')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Get the count of each category and sort in descending order
category_counts = df['category_3'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.countplot(x='category_3', data=df, order=category_counts.index, palette='viridis')
plt.title('Distribution of Category 3')
plt.xlabel('Category')
plt.ylabel('Count')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

### Average, Minimum and Maximum Rating of category


```python
result1 = df.groupby('category_2')['rating'].agg(['mean','min','max'])
result1['mean'] = result1['mean'].round(2)
result1
```

```python
# Plotting line plot for min, mean, and max ratings
result1[['min', 'mean', 'max']].plot(kind='line', marker='o', figsize=(10, 5))

# Set titles and labels
plt.title('Ratings Summary by Product Category_2')
plt.xlabel('Product Category')
plt.ylabel('Avg, min, max rating')
plt.legend()
plt.xticks(range(len(result1.index)), result1.index, rotation=45, ha='right')  # Use the index as x-axis labels
plt.tight_layout()  # Adjust layout to prevent clipping
# Show the plots
plt.show()
```
